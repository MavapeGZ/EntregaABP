```python
import pandas as pd
```

```python
path="./data/merged_data.csv"
originalData=pd.read_csv(path)
## Elimina los duplicados
originalData = originalData.drop_duplicates(subset='Title')
```

```python
originalData.info()
```

```python
from nltk.tokenize import word_tokenize 
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer

import nltk
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('punkt_tab') 
##nltk.download()

ps = PorterStemmer()

preprocessedText = []

for row in originalData.itertuples():
    
    
    text = word_tokenize(str(row[6])) ## indice de la columna que contiene el texto
    ## Remove stop words
    stops = set(stopwords.words("english"))
    text = [ps.stem(w) for w in text if not w in stops and w.isalnum()]
    text = " ".join(text)
    
    preprocessedText.append(text)

preprocessedData = originalData
preprocessedData['processed_text'] = preprocessedText

preprocessedData
```

```python
filteredData = preprocessedData.loc[:, ['Title', 'processed_text', 'Developer', 'Publisher', 'Popular Tags', 'Game Features']]
filteredData.head(10)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

bagOfWordsModel = TfidfVectorizer()
bagOfWordsModel.fit(filteredData['processed_text'])
textsBoW= bagOfWordsModel.transform(filteredData['processed_text'])
print("Finished")
```

```python
from sklearn.metrics import pairwise_distances

distance_matrix= pairwise_distances(textsBoW, metric='cosine')
```

```python
print(distance_matrix.shape)
print(type(distance_matrix))
```

```python
searchTitle = "Call of Duty¬Æ: Black Ops II" #Pel√≠cula base para las recomendaciones
indexOfTitle = preprocessedData[preprocessedData['Title']==searchTitle].index.values[0]
indexOfTitle
```

```python
distance_scores = list(enumerate(distance_matrix[indexOfTitle]))
ordered_scores = sorted(distance_scores, key=lambda x: x[1])
top_scores = ordered_scores[1:11]
top_indexes = [i[0] for i in top_scores]
```
